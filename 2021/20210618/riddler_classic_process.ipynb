### Capturing Possible Crashes

The testing below is to determine how to best handle determining crash or not. 

```python
def test_seq(car_1, car_2):
    """Testing some scenarios"""
    # modify so that we always consider clockwise, for easier test
    o_a, e_a = min(car_1), max(car_1)
    o_b, e_b = min(car_2), max(car_2)
    #print(o_a, e_a)
    #print(o_b, e_b)
    # angle within - no crash 
    if o_a < o_b < e_a and o_a < e_b < e_a:
        return 0
    
    # route is totally outside 
    if (not(o_a <= o_b <= e_a)) and (not(o_a <= e_b <= e_a)):
        return 0
    
    # 
    
    return 1
    
# second car is between roads 
assert(test_seq((1,8), (2,3)) == 0)

# second car would cross here since going to same location 
assert(test_seq((1,3), (4,3)) == 1)

# no crash as they are totally separate
assert(test_seq((1,3), (8,5)) == 0)
```
